```python
import pandas as pd
import subprocess
import os
%matplotlib inline
import numpy as np

import matplotlib as mpl
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
url="https://covid.ourworldindata.org/data/owid-covid-data.csv"  #Extract data from URL
data = pd.read_csv(url) #Reading extracted data from URL
data.head() #Display first 5 rows of the data
```

```python
GER = data[data['location'] == 'Germany'].reset_index(drop=True)  #Extract data of three countries in three variables
CZE = data[data['location'] == 'Czechia'].reset_index(drop=True)
FRA = data[data['location'] == 'France'].reset_index(drop=True)
```

```python
GER['cases_per_population'] = GER['total_cases']/GER['population'] #Calculate the absolute Covid cases/population
GER ['Vaccine_percentage'] = GER['people_vaccinated']*100/GER['population'] #Calculate the vaccination percentage
```

```python
CZE['cases_per_population'] = CZE['total_cases']/CZE['population'] #Calculate the absolute Covid cases/population
CZE['Vaccine_percentage'] = CZE['people_vaccinated']*100/CZE['population'] #Calculate the vaccination percentage
```

```python
FRA['cases_per_population'] = FRA['total_cases']/FRA['population'] #Calculate the absolute Covid cases/population
FRA ['Vaccine_percentage'] = FRA['people_vaccinated']*100/FRA['population'] #Calculate the vaccination percentage
```

```python
combined_list = pd.concat([GER,CZE,FRA]) #Make combined list of all countries to display
```

```python
fig = px.line(combined_list,x = "date",y = "cases_per_population", title = "COVID cases", color = 'location') #Plotting
fig.show()
```

```python
fig = px.line(combined_list,x = "date",y = "Vaccine_percentage", title = "COVID vaccination percentage", color = 'location')
fig.show()
```
